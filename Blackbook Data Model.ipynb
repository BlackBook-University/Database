```python
def clean_fall2010_data(database):
    database=database.iloc[:,:29]
    gender_array = database['Gender']
    Age_group = database['Age Group']
    ethnicity = database['Ethnicity']
    basic_enrollment = database['Fall 2010']
    basic_retention = database['Fall 2010.1']
    basic_success = database['Fall 2010.2']
    basic_retrate = database['Fall 2010.3']
    basic_succrate = database['Fall 2010.4']
    combined = {'Gender':gender_array, 'Age Group':Age_group, 'Ethnicity':ethnicity, 'Enrollment Count':basic_enrollment,
               'Retention Count':basic_retention, 'Success Count':basic_success, 'Retention Rate':basic_retrate, 'Success Rate':basic_succrate}
    fall2010 = pd.DataFrame(data=combined)
    fall2010 = fall2010.drop([0,1])
    return fall2010

clean_fall2010_data(start_cleaning)
```
